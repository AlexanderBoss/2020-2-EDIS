```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G =nx.Graph()
```

Los nodos

```python
G.add_node(1)
G.add_node(2)
G.add_node(3)
```

Conexiones de las aristas

```python
G.add_edge(1,2)
G.add_edge(2,3)
G.add_edge(1,3)
```

Impresion de los nodos

```python
print(G.nodes)
```

```python
print(G.edges)
```

```python
nx.draw(G)
plt.show()
```
